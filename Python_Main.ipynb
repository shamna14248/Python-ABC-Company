```python
#1. Preprocessing
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('/content/myexcel - myexcel.csv.csv')  # replace with your file name

# Replace 'Height' with random values between 150 and 180
df['Height'] = np.random.randint(150, 181, size=len(df))
```

```python
#2. Analysis Tasks
# (1) Team-wise Employee Distribution & Percentage
team_counts = df['Team'].value_counts()
team_percent = (team_counts / len(df)) * 100
print("Team-wise Employee Count:\n", team_counts)
print("\nTeam-wise Percentage Distribution:\n", team_percent)

# (2) Position-wise Employee Segregation
position_counts = df['Position'].value_counts()
print("\nPosition-wise Employee Count:\n", position_counts)

# (3) Predominant Age Group
age_bins = [0, 25, 30, 35, 40, 100]
age_labels = ['<25', '25-30', '30-35', '35-40', '40+']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
age_group_counts = df['Age Group'].value_counts()
print("\nAge Group Distribution:\n", age_group_counts)
print("\nMost Common Age Group:", age_group_counts.idxmax())

# (4) Team & Position with Highest Salary Expenditure
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)
print("\nTotal Salary by Team:\n", team_salary)
print("\nTeam with Highest Salary Expenditure:", team_salary.idxmax())

print("\nTotal Salary by Position:\n", position_salary)
print("Position with Highest Salary Expenditure:", position_salary.idxmax())

# (5) Correlation Between Age and Salary
correlation = df['Age'].corr(df['Salary'])
print("\nCorrelation between Age and Salary:", correlation)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Team distribution bar chart
team_counts.plot(kind='bar', title='Employees per Team')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.show()

# 2. Team percentage pie chart
team_percent.plot(kind='pie', autopct='%1.1f%%', title='Team Distribution Percentage')
plt.ylabel('')
plt.show()

# 3. Position count bar chart
position_counts.plot(kind='bar', title='Employees by Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.show()

# 4. Age group histogram
age_group_counts.plot(kind='bar', title='Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()

# 5. Age vs Salary scatter plot
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Correlation Between Age and Salary')
plt.show()
```
